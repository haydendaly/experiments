```python
# Import Packages
%matplotlib inline
%matplotlib notebook
import zmq
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
# Connect to Publisher
context = zmq.Context()
socket = context.socket(zmq.SUB)
socket.connect("tcp://localhost:5555")
socket.subscribe('11098000')

# Initialize Figure
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
fig.canvas.draw()
x_axis = []
y_axis = []

while True:
    # Get Data from Publisher
    site_no = socket.recv_string()
    datetime = socket.recv_string()
    discharge = socket.recv_string()
    qualification = socket.recv_string()
    
    x_axis.append(datetime)
    y_axis.append(float(discharge))
    ax.clear()
    ax.plot(x_axis, y_axis, label=site_no)
    fig.canvas.draw()
```
