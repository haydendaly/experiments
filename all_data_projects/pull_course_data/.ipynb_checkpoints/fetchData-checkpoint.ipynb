```python
import requests
import json
import re
url = 'https://stevens-scheduler.cfapps.io/p/'
```

```python
def refreshData():
    term = requests.get(url + 'terms').json()[0]
    res = requests.get(url + term)
    temp = []
    courses = {}
    for course in res.json():
        try:
            code = course['section'].split()
            key = code[0]
            number = code[1][0:3]
            if not (key + number) in courses:
                courses[key + number] = 1
                temp.append({
                    "className" : course["title"],
                    "key" : key + " " + number
                })
        except Exception as e:
            print(e)
    print(courses)
    return temp
        
data = refreshData()
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
def getData():
    f = open('data.json')
    return json.load(f)
        
data = getData()
```
