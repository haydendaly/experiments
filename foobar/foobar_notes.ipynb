```python
from fractions import Fraction
from itertools import starmap
from operator import mul

def multiply(A, B):
    return [[sum(starmap(mul, zip(row, col))) for col in zip(*B)] for row in A]

def elim(row_A, row_B, col, val):
    ol = (row_B[col] - val) / row_A[col]
    for i in range(len(row_B)):
        row_B[i] -= ol * row_A[i]
        
def gcd(a, b):
    if b == 0:
        return a
    while b:
        a, b = b, a % b
    return a

def gauss_jordan(A):
    for i in range(len(A)):
        if A[i][i] == 0:
            for j in range(i+1, len(A)):
                if A[i][j] != 0:
                    A[i], A[j] = A[j], A[i]
                    break
        for j in range(i+1, len(A)):
            elim(A[i], A[j], i, 0)
    for i in range(len(A)-1, -1, -1):
        for j in range(i-1, -1, -1):
            elim(A[i], A[j], i, 0)
    for i in range(len(A)):
        elim(A[i], A[i], i, 1)
    return A

def inverse(A):
    temp = [[] for _ in A]
    for i, row in enumerate(A):
        temp[i].extend(row + [0]*i + [1] + [0]*(len(A)-i-1))
    gauss_jordan(temp)
    result = []
    for i in range(len(temp)):
        result.append(temp[i][len(temp[i])//2:])
    return result

def solution(m):
    l = len(m)
    if l == 0:
        return []
    elif l < 2:
        return [1, 1]
    matrix = [[0]*l for _ in m]
    denominators = [sum(row) for row in m]
    indices = dict()
    idx = 0
    for i, row in enumerate(m):
        if sum(row) == 0:
            indices[i] = idx
            matrix[idx][idx] = Fraction(1,1)
            idx += 1
    identity_end = idx
    for i, row in enumerate(m):
        if i not in indices:
            indices[i] = idx
            idx += 1
    for i, row in enumerate(m):
        for j, numerator in enumerate(row):
            if numerator != 0:
                denominator = denominators[i]
                matrix[indices[i]][indices[j]] = float(numerator) / float(denominators[i])
    cycles = [row[identity_end:] for row in matrix[identity_end:]]
    
    l = len(cycles)
    for i in range(l):
        for j in range(l):
            cycles[i][j] *= -1
            if i == j:
                cycles[i][j] += 1
    F = inverse(cycles)
    remainder = [row[:identity_end] for row in matrix[identity_end:]]
    
    new_indices = dict()
    for key in indices:
        new_indices[indices[key]] = key 
    
    result = []
    FR = multiply(F, remainder)
    for i, row in enumerate(FR):
        if denominators[new_indices[identity_end + i]]:
            result = [Fraction(num).limit_denominator() for num in row]
            break

    den = 1
    for frac in result:
        den = den // gcd(den, frac.denominator) * frac.denominator
    ans = []
    for frac in result:
        ans.append(frac.numerator * den // frac.denominator)
    ans.append(den)
    return ans
```

```python
from fractions import gcd

def build_graph(arr):
    l = len(arr)
    def has_cycle(i, j):
        if i == j:
            return False
        num = (arr[i] + arr[j]) // gcd(arr[i], arr[j])
        return (num & (num - 1)) != 0 # check if not power of 2
    # can't use defaultdict(list) bc need to iterate over later
    graph = dict()
    for i in range(l):
        graph[i] = []
    for i in range(l):
        for j in range(l):
            if has_cycle(i, j):
                graph[i].append(j)
    return graph

# modelled after https://stanford.edu/~rezab/classes/cme323/S16/projects_reports/shoemaker_vare.pdf
# can't use traditional matching algs bc not bipartite, sorry cs 600 :-/
def blossalg(graph):
    nodes, edges, valid = [], dict(), True
    for i in graph:
        nodes.append(i)
        for j in graph[i]:
            edges[(i, j)] = False
    while valid:
        valid = False
        for node in nodes:
            a, path, seen, check = node, [], set(), False
            while True:
                seen.add(a)
                for b in graph[a]:
                    if not (b in seen) and not (check ^ edges[(a, b)]):
                        path.extend((a, b))
                        break
                if len(path) == 0:
                    break
                if path[-1] in nodes and path[-1] != node:
                    for i in range(len(path) - 1):
                        new_val = not edges[(path[i], path[i + 1])]
                        edges[(path[i], path[i + 1])] = new_val
                        edges[(path[i + 1], path[i])] = new_val
                    # remove pair
                    nodes.remove(path[0])
                    nodes.remove(path[-1])
                    valid = True # continue iteration
                    break
                a = path.pop()
                check = not check
    return nodes

def solution(arr):
    graph = build_graph(arr)
    return len(blossalg(graph))

assert(solution([1,1]) == 2)
assert(solution([1, 7, 3, 21, 13, 19]) == 0)
assert(solution([1]) == 1)
assert(solution([1, 7, 1, 1]) == 4)
```

```python
from copy import deepcopy

def bellman_ford(source, times, distance, m):    
    # psuedo code from https://en.wikipedia.org/wiki/Bellman%E2%80%93Ford_algorithm
    # could use https://en.wikipedia.org/wiki/Shortest_Path_Faster_Algorithm to speed up
    distance[source][source] = 0 # should be diagonal of zeroes
    for _ in range(m - 1):
        for i in range(m):
            for j in range(m):
                if distance[source][i] !=  float('inf') and distance[source][i] + times[i][j] < distance[source][j]:
                    distance[source][j] = distance[source][i] + times[i][j]
    
    for i in range(m):
        for j in range(m):
            if distance[source][i] != float('inf') and distance[source][i] + times[i][j] < distance[source][j]:
                return False # contains cycle
    return True
         
def has_path(times, distance, m):
    for source in range(m):
        if not bellman_ford(source, times, distance, m):                
            return False
    return True

def find_paths(source, target, t, distance, m):
    adj_list = [[i] for i in range(m)]
    s = [(source, t, [source], 
            adj_list)]
    all_nodes = set(range(m))
    result = []
    end_reached = False
    
    while s and not end_reached:
        (node, t, path, adj_list) = s.pop()
        for adj_node in all_nodes - set(adj_list[node]):
            next_adj_list = deepcopy(adj_list)
            # if they negate one another, add to next adjancency list
            if  distance[adj_node][node] == -distance[node][adj_node]:
                next_adj_list[node].append(adj_node)
            if (t - distance[node][adj_node] - distance[adj_node][target] >= 0): 
                path_next = path + [adj_node]    
                s.append((adj_node, t - distance[node][adj_node], path_next, next_adj_list))          
                if adj_node == target:
                    result.append(set(path_next))
                    if len(result[-1]) == m:
                        end_reached = True
                        break
    return result
        
def solution(times, time_limit):
    m = len(times)
    if m <= 2:
        return [] # no bunnies
    distance = [[float('inf') if i != j else 0 for j in range(m)] for i in range(m)]
    max_free = set()
    is_viable = has_path(times, distance, m)
    
    if not is_viable:
        return range(m - 2)
    
    paths = find_paths(0, m - 1, time_limit, distance, m)
    for free in paths:
        if len(max_free) < len(free) or (len(max_free) == len(free) and sum(max_free) > sum(free)):
            max_free = free        

    result = []
    for bunny in max_free:
        if bunny != 0 and bunny != m - 1: # don't account for self or end state
            result.append(bunny - 1)
    result.sort()
    return result
    
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def exponents(row, col):
    result = 0
    for r in row:
        for c in col:
            result += gcd(r, c)
    return result
            
def cycles(nums, n):
    result = factorial(n)
    count = dict()
    for num in nums:
        if num not in count:
            count[num] = 0
        count[num] += 1
    for num in count:
        result = result // ((num ** count[num]) * factorial(count[num]))
    return result

# accelerated partioning algorithm from this blog: 
# https://jeromekelleher.net/category/combinatorics.html
def integer_partitions(n):
    a = list(range(n + 1))
    k, y = 1, n - 1
    while k:
        x = a[k - 1] + 1
        k -= 1
        while 2 * x <= y:
            a[k] = x
            y -= x
            k += 1
        l = k + 1
        while x <= y:
            a[k], a[l] = x, y
            yield a[:k + 2]
            x += 1
            y -= 1
        a[k] = x + y
        y = x + y - 1
        yield a[:k + 1]
        
def factorial(num):
    if num <= 1:
        return 1
    return num * factorial(num - 1)

def solution(w, h, s):    
    # uses burnside's lemma from abstract algebra (thanks again ma 334)
    result = 0
    for row in integer_partitions(w):
        for col in integer_partitions(h):     
            m = cycles(row, w) * cycles(col, h)
            result += m * (s ** exponents(row, col))
    return result // (factorial(w) * factorial(h))
    
FACTORIAL_CACHE = [
    1,
    2,
    6,
    24,
    120,
    720,
    5040,
    40320,
    362880,
    3628800,
    39916800,
    479001600,
    6227020800,
    87178291200,
    1307674368000,
    20922789888000,
    355687428096000,
    6402373705728000,
    121645100408832000,
    2432902008176640000,
    51090942171709440000,
    1124000727777607680000,
    25852016738884976640000,
    620448401733239439360000,
    15511210043330985984000000,
    403291461126605635584000000,
    10888869450418352160768000000,
    304888344611713860501504000000,
    8841761993739701954543616000000,
    265252859812191058636308480000000,
    8222838654177922817725562880000000,
    263130836933693530167218012160000000,
    8683317618811886495518194401280000000,
    295232799039604140847618609643520000000,
    10333147966386144929666651337523200000000,
    371993326789901217467999448150835200000000,
    13763753091226345046315979581580902400000000,
    523022617466601111760007224100074291200000000,
    20397882081197443358640281739902897356800000000
]

print(solution(2, 2, 2)) # 7
print(solution(2, 3, 4)) #430
```
